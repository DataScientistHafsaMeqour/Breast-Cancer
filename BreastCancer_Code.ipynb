```python
X=df.iloc[:,2:31].values
Y=df.iloc[:,1].values
```

```python
from sklearn.model_selection import train_test_split
```

```python
# Séparation des données en ensembles d'entraînement et de test
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.25,random_state=42)
```

```python
#utilisée pour mettre à l'échelle les colonnes
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
X_train=sc.fit_transform(X_train)
X_test=sc.fit_transform(X_test)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier


#Création et entraînement de modèles de classification 
def model(X_train,Y_train):
    model=LogisticRegression()
    model.fit(X_train,Y_train)
    
    model2=DecisionTreeClassifier(criterion='entropy')
    model2.fit(X_train,Y_train)

    model3=RandomForestClassifier(n_estimators=10,criterion='entropy')
    model3.fit(X_train,Y_train)
    #Évaluation des modèles sur les données d'entraînement
    print('[0] LogisticRegression : ',model.score(X_train,Y_train))
    print('[1] DecisionTree : ',model2.score(X_train,Y_train))
    print('[2] Random Forest : ',model3.score(X_train,Y_train))
    return model,model2,model3
    
```

```python
model=model(X_train,Y_train)
```

```python
#évaluer les performances de plusieurs modèles en calculant la matrice de confusion

from sklearn.metrics import confusion_matrix
for i in range(len(model)):
    print('Model ',i+1)
    cm=confusion_matrix(Y_test,model[i].predict(X_test))
    print(cm)
    TP = cm[0][0]
    FP = cm[0][1]
    FN = cm[1][0]
    TN = cm[1][1]
    print('Accuracy =',(TP+TN)/(TP+TN+FP+FN))
```
